```python
class Solution:
    def maximalRectangle(self, matrix):
        max_area = 0
        for row in range(0, len(matrix)):
            for col in range(0, len(matrix[0])):
                #Bottom
                for endrow in range(row, len(matrix)):
                    for endcol in range(col, len(matrix[0])):
                        rectangle = [x[col:endcol+1] for x in matrix[row:endrow+1]]
                        if self.is_homogenuous(rectangle):
                            #print(rectangle, (row, col), (endrow, endcol))
                            max_area = max(max_area, (endrow+1-row) * (endcol+1-col))
                
        return max_area

    def is_homogenuous(self, rectangle):
        for row in rectangle:
            for val in row:
                if val != "1":
                    return False
        return True
```
